#NoBackOffValueID

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df2 = pd.read_csv(io.BytesIO(uploaded['NoBackOffValeurID.csv']))
```

```python
tab = np.zeros((51,50,3),int)

for i in range(51):
  for j in range(50):
    tab[i][j][0]= df2["time"][i*50+j]
    tab[i][j][1]= df2["round"][i*50+j]
    tab[i][j][2]= df2["nbMessages"][i*50+j]

time = [tab[i][j][0] for i in range(51) for j in range(50)]
round = [tab[i][j][1] for i in range(51) for j in range(50)]
msg = [tab[i][j][2] for i in range(51) for j in range(50)]
```

```python
avg_time = []
avg_round = [] 
avg_msg = []
noeud = []
for i in range(51):
  val_time = 0
  val_round = 0
  val_msg = 0

  for j in range(50):
    val_time += time[i*50+j]
    val_round += round[i*50+j]
    val_msg += msg[i*50+j]

  avg_time.append(val_time/50.0)
  avg_round.append(val_round/50.0)
  avg_msg.append(val_msg/50.0)
  noeud.append(10+10*i)
```

```python
fig = plt.figure(figsize=(30, 15))
(ax1, ax2), (ax3, ax4) = fig.subplots(2,2)

ax1.plot(noeud, avg_time, label = "Time", color = "blue")
ax1.set_xlabel("Nombre de noeuds N")
ax1.set_ylabel("Temps")
ax1.legend(loc = "upper left")
ax1.set_title("Temps en fonction du nombre de noeuds N")

ax2.plot(noeud, avg_round, label = "Round", color = "red")
ax2.set_xlabel("Nombre de noeuds N")
ax2.set_ylabel("Round")
ax2.legend(loc = "upper left")
ax2.set_title("Nombre de round en fonction du nombre de noeuds N")

ax3.plot(noeud, avg_msg, label = "Messages", color = "black")
ax3.set_xlabel("Nombre de noeuds N")
ax3.set_ylabel("Nombre de messages")
ax3.legend(loc = "upper left")
ax3.set_title("Nombre de messages en fonction du nombre de noeuds N")

ax4.plot(noeud, avg_round, label = "Round", color = "red")
ax4.plot(noeud, avg_time, label = "Time", color = "blue")
ax4.plot(noeud, avg_msg, label = "Messages",  color = "black")
ax4.set_xlabel("Nombre de noeuds N")
ax4.legend(loc = "upper left")
ax4.set_title("MÃ©triques en fonction du nombre de noeuds N")

plt.show()
```
